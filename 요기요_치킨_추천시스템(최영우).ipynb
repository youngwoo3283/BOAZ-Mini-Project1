```python
import numpy as np
from sklearn.metrics import mean_squared_error

def get_rmse(R, P, Q, non_zeros):
    error = 0
    # ë‘ê°œì˜ ë¶„í•´ëœ í–‰ë ¬ Pì™€ Q.Tì˜ ë‚´ì  ê³±ìœ¼ë¡œ ì˜ˆì¸¡ R í–‰ë ¬ ìƒì„±
    full_pred_matrix = np.dot(P, Q.T)
    
    # ì‹¤ì œ R í–‰ë ¬ì—ì„œ ë„ì´ ì•„ë‹Œ ê°’ì˜ ìœ„ì¹˜ ì¸ë±ìŠ¤ ì¶”ì¶œí•˜ì—¬ ì‹¤ì œ R í–‰ë ¬ê³¼ ì˜ˆì¸¡ í–‰ë ¬ì˜ RMSE ì¶”ì¶œ
    x_non_zero_ind = [non_zero[0] for non_zero in non_zeros]
    y_non_zero_ind = [non_zero[1] for non_zero in non_zeros]
    R_non_zeros = R[x_non_zero_ind, y_non_zero_ind]
    
    full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind]
      
    mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
    rmse = np.sqrt(mse)
    
    return rmse
```

```python
#Rì€ ê¸°ì¡´ í‰ì  í–‰ë ¬
#KëŠ” P,Që¥¼ ì´ˆê¸°í™”í•˜ê¸° ìœ„í•œ ìƒìˆ˜


def matrix_factorization(R, K, steps=200, learning_rate=0.01, r_lambda = 0.01):
    num_users, num_items = R.shape
    # Pì™€ Q ë§¤íŠ¸ë¦­ìŠ¤ì˜ í¬ê¸°ë¥¼ ì§€ì •í•˜ê³  ì •ê·œë¶„í¬ë¥¼ ê°€ì§„ ëœë¤í•œ ê°’ìœ¼ë¡œ ì…ë ¥í•©ë‹ˆë‹¤. 
    np.random.seed(1)
    P = np.random.normal(scale=1./K, size=(num_users, K))
    Q = np.random.normal(scale=1./K, size=(num_items, K))

    break_count = 0
       
    # R > 0 ì¸ í–‰ ìœ„ì¹˜, ì—´ ìœ„ì¹˜, ê°’ì„ non_zeros ë¦¬ìŠ¤íŠ¸ ê°ì²´ì— ì €ì¥. 
    # NON-ZERO: ì‚¬ìš©ìê°€ ë¨¹ì€ ì¹˜í‚¨ë“¤ë§Œ ë”°ë¡œ í–‰ë ¬ì„ ë§Œë“¤ìŒ
    non_zeros = [ (i, j, R[i,j]) for i in range(num_users) for j in range(num_items) if R[i,j] > 0 ]
   
    # SGDê¸°ë²•ìœ¼ë¡œ Pì™€ Q ë§¤íŠ¸ë¦­ìŠ¤ë¥¼ ê³„ì† ì—…ë°ì´íŠ¸. 
    for step in range(steps):
        for i, j, r in non_zeros:
            # ì‹¤ì œ ê°’ê³¼ ì˜ˆì¸¡ ê°’ì˜ ì°¨ì´ì¸ ì˜¤ë¥˜ ê°’ êµ¬í•¨
            eij = r - np.dot(P[i, :], Q[j, :].T)
            # Regularizationì„ ë°˜ì˜í•œ SGD ì—…ë°ì´íŠ¸ ê³µì‹ ì ìš©
            P[i,:] = P[i,:] + learning_rate*(eij * Q[j, :] - r_lambda*P[i,:])
            Q[j,:] = Q[j,:] + learning_rate*(eij * P[i, :] - r_lambda*Q[j,:])
       
        rmse = get_rmse(R, P, Q, non_zeros)
        if (step % 10) == 0 :
            print("### iteration step : ", step," rmse : ", rmse)
            
    return P, Q
```

```python
def get_unate_surprise(ratings, userId):
    #ì…ë ¥ê°’ìœ¼ë¡œ ë“¤ì–´ì˜¨ userIdì— í•´ë‹¹í•˜ëŠ” ì‚¬ìš©ìê°€ í‰ì ì„ ë§¤ê¸´ ëª¨ë“  ì¹˜í‚¨ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±
    # seen_movies = ratings[ratings['userId']== userId]['movieId'].tolist()
    ate_chickens = ratings[ratings['userID']== userId]['itemID'].tolist()
    
    # ëª¨ë“  ì˜í™”ë“¤ì˜ movieIdë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±. 
    #total_movies = movies['movieId'].tolist()
    total_food = ratings['itemID'].unique().tolist()

    # ëª¨ë“  ì¹˜í‚¨ë“¤ì¤‘ ì´ë¯¸ í‰ì ì„ ë§¤ê¸´ ì¹˜í‚¨ì„ ì œì™¸í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±
    #unseen_movies= [movie for movie in total_movies if movie not in seen_movies]
    unate_chickens= [food for food in total_food if food not in ate_chickens]
    
    print('í‰ì  ë§¤ê¸´ ì¹˜í‚¨ìˆ˜:',len(ate_chickens), 'ì¶”ì²œëŒ€ìƒ ì¹˜í‚¨ìˆ˜:',len(unate_chickens), \
          'ì „ì²´ ì¹˜í‚¨ìˆ˜:',len(total_food))
    
    return unate_chickens

unate_chickens = get_unate_surprise(df_sus, 'sh')
```

```python
def recomm_chicken_by_surprise(algo, userId, unate_chickens, top_n=10):
    # ì•Œê³ ë¦¬ì¦˜ ê°ì²´ì˜ predict() ë©”ì„œë“œë¥¼ í‰ì ì´ ì—†ëŠ” ì¹˜í‚¨ì— ë°˜ë³µ ìˆ˜í–‰í•œ í›„ ê²°ê³¼ë¥¼ list ê°ì²´ë¡œ ì €ì¥
    predictions = [algo.predict(str(userId), str(chicken_name)) for chicken_name in unate_chickens]
    
    # predictions list ê°ì²´ëŠ” surpriseì˜ Predictions ê°ì²´ë¥¼ ì›ì†Œë¡œ ê°€ì§€ê³  ìˆìŒ.
    # [Prediction(uid='9', iid='1', est=3.69), Prediction(uid='9', iid='2', est=2.98),,,,]
    # ì´ë¥¼ est ê°’ìœ¼ë¡œ ì •ë ¬í•˜ê¸° ìœ„í•´ì„œ ì•„ë˜ì˜ sortkey_est í•¨ìˆ˜ë¥¼ ì •ì˜í•¨.
    # sortkey_est í•¨ìˆ˜ëŠ” list ê°ì²´ì˜ sort() í•¨ìˆ˜ì˜ í‚¤ ê°’ìœ¼ë¡œ ì‚¬ìš©ë˜ì–´ ì •ë ¬ ìˆ˜í–‰.
    def sortkey_est(pred):
        return pred.est
    
    # sortkey_est( ) ë°˜í™˜ê°’ì˜ ë‚´ë¦¼ ì°¨ìˆœìœ¼ë¡œ ì •ë ¬ ìˆ˜í–‰í•˜ê³  top_nê°œì˜ ìµœìƒìœ„ ê°’ ì¶”ì¶œ.
    predictions.sort(key=sortkey_est, reverse=True)
    top_predictions= predictions[:top_n]
    
    # top_nìœ¼ë¡œ ì¶”ì¶œëœ ì˜í™”ì˜ ì •ë³´ ì¶”ì¶œ. ì˜í™” ì•„ì´ë””, ì¶”ì²œ ì˜ˆìƒ í‰ì , ì œëª© ì¶”ì¶œ
    # top_movie_ids = [ int(pred.iid) for pred in top_predictions]
    # top_movie_rating = [ pred.est for pred in top_predictions]
    # top_movie_titles = movies[movies.movieId.isin(top_movie_ids)]['title']
    # top_movie_preds = [ (id, title, rating) for id, title, rating in zip(top_movie_ids, top_movie_titles, top_movie_rating)]
    
    return top_predictions

unate_chickens = get_unate_surprise(df_sus, 'ck')
top_chicken_preds = recomm_chicken_by_surprise(algo, 'ck', unate_chickens, top_n=10)
print('##### Top-10 ì¶”ì²œ ì˜í™” ë¦¬ìŠ¤íŠ¸ #####')

for top_rem in top_chicken_preds:
    #print(top_movie[1], ":", top_movie[2])
    print(top_rem[1])
```
